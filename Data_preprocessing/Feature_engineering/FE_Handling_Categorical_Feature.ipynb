#### Onehotencoding with many categories in a feature

```python
df=pd.read_csv('mercedesbenz.csv',usecols=["X0","X1","X2","X3","X4","X5","X6"])
df.head()
```

```python
for i in df.columns:
    print(len(df[i].unique()))
```

```python
df.X1.value_counts().sort_values(ascending=False).head(10)
# df.X1.value_counts().sort_values(ascending=False).head(10).index - will give us the index value
```

```python
lst_10=df.X1.value_counts().sort_values(ascending=False).head(10).index
lst_10=list(lst_10)  #first it intype index so we chaneged into list
```

```python
lst_10
```

```python
import numpy as np
for categories in lst_10:
    df[categories]=np.where(df['X1']==categories,1,0)
```

```python
# df[['X1'],[lst_10]] -willnot work

# lst_10.append('X1')
lst_10.insert(0, 'X1')
```
