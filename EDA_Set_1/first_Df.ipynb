```python
df_country=pd.read_excel('Zomato/Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
final_df
#how -left join or right join
# we used left join ,here all the rows will get printed and only matched values in col o f2nd df will get print 
# if we used right join ,here all the rows og right df will get printed and only matched values in col o 1st df will get print 
```

```python
country_names=final_df.Country.value_counts().index
country_names
# .index will give us each country name
```

```python
country_val=final_df.Country.value_counts().values
country_val
# it just printed the totao value of th each country
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.1f%%")
plt.show()
# [:3] because of this only till 3 value will get printed
```

Observation:Zomato maximum records or transaction are from India After that USA and then United Kingdoms
