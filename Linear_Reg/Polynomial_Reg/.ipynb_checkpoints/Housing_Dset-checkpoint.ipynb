```python
from sklearn.preprocessing import PolynomialFeatures

# Transform the independent variable to include polynomial features
degree_of_poly = 2
poly = PolynomialFeatures(degree=degree_of_poly)
X_train_poly = poly.fit_transform(X_train.reshape(-1, 1))
X_test_poly = poly.transform(X_test.reshape(-1, 1))

# This is because the fit_transform method expects a 2D array-like object as input, and without the reshape operation, the arrays will remain 1D.

# The PolynomialFeatures class in scikit-learn expects a 2D array-like input, where each row represents a data point, and each column represents a feature. If you pass a 1D array without reshaping, it will be interpreted as a single sample with multiple features, which can lead to unexpected behavior or errors.
```

```python
# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_train_poly, y_train)
y_pred = model.predict(X_test_poly)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse_poly = mean_squared_error(y_test, y_pred)
r2_poly = r2_score(y_test, y_pred)

print("Polynomial Regression (Degree = 2) - Mean Squared Error:", mse_poly)
print("Polynomial Regression (Degree = 2) - R-squared:", r2_poly)

# Polynomial Regression Coefficients: [ 0.00000000e+00 -1.72150574e-02  1.46402648e-05]
# Polynomial Degree = 2 (Linear Regression) so we have 3 coef,x0,x1,x2
```
