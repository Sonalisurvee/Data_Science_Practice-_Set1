```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]
var_2 = [7,3,4,1,5,3,4]


df = pd.DataFrame({'var':var,'var_1':var_1,'var3':var_2})

sns.lineplot(x='var',y='var_1',data=df)

plt.show()
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
#3D PLOT
ax = plt.axes(projection="3d")
z = np.linspace(0,30,100)
x = np.sin(z)
y = np.cos(z)
ax.plot3D(x,y,z)
plt.show()
```
