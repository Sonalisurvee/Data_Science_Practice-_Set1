```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

iris = load_iris()
X = iris.data  # Features
y = iris.target  # Target (labels, for supervised tasks)
```

```python
# Standardize the features (optional but recommended for K-means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
num_clusters = 3  # Number of clusters to create (since there are 3 classes in Iris dataset)
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X_scaled)

# Get cluster centers and labels
cluster_centers = kmeans.cluster_centers_
cluster_labels = kmeans.labels_
```

```python
# Print cluster centers to interpret cluster characteristics
print("Cluster Centers (Centroids):")
print(cluster_centers)
print("Cluster labels:")
print(cluster_labels)

# [[ 1.13597027  0.08842168  0.99615451  1.01752612]
# the Iris dataset has four features, so each cluster center has four values representing the mean value of the data points within that cluster along each feature dimension.
```

```python
# Visualize the clusters in 2D (using the first two features for simplicity)
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k', s=50)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='X', s=200, label='Cluster Centers')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.title('K-means Clustering - Iris Dataset')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(X_scaled, cluster_labels)
print("Silhouette Score:", silhouette_avg)
```

```python
# Assuming 'new_data' is a new set of data points (e.g., a 2x4 array representing 2 new data points)
new_data = np.array([[5.1, 3.5, 1.4, 0.2],
                     [7.0, 3.2, 4.7, 1.4]])

# Assuming 'new_data' is a new set of data points
new_data_scaled = scaler.transform(new_data)
new_cluster_labels = kmeans.predict(new_data_scaled)
print("Cluster Labels for New Data Points:")
print(new_cluster_labels)
```

```python
# Calculate the variance explained for different K values
variance_explained = []
K_values = range(2, 11)

for K in K_values:
    kmeans = KMeans(n_clusters=K, random_state=42)
    kmeans.fit(X_scaled)
    variance_explained.append(kmeans.inertia_)

# Plot the variance explained vs. K values
plt.figure(figsize=(8, 6))
plt.plot(K_values, variance_explained, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Variance Explained')
plt.title('Elbow Method for Optimal K')
plt.show()
```
